*What is uses of EDA?*

Exploratory Data Analysis (EDA) is an approach to analyze the data using visual techniques.
It is used to discover trends, patterns, or to check assumptions with the help of statistical summary and graphical representations.

#Graphical analysis

```python
df['continent'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

**a vast majority of the continent is “Asia” in this dataset**

```python
df['case_status'].value_counts().plot(kind="pie", autopct="%.2f")
```

 **a vast majority of the case status is “Certified” in this dataset**

```python
sns.displot(df, x="prevailing_wage")
```

```python
sns.pairplot(df)
```

```python
sns.distplot(np.log(df['prevailing_wage']))
```

```python
sns.barplot(df['prevailing_wage'], df['education_of_employee'])
```

```python
df_certified = df[(df["case_status"]=='Certified')]
df_certified
```

```python
df_certified.groupby('yr_of_estab').count().plot() 
#Has there been an increase or decrease in the number of visas granted over the years
```

**We can see from the above graph that the number of  visas being issued has been increasing over the years**

```python
df_certified[df_certified["full_time_position"] == "Y"].count() 
#having a full time job position influence your chance of landing the visa
```

```python
df_certified["full_time_position"].hist() #The histogram tells us that people having visas majorly work in full time positions.
```

```python
#would be is there are trend in the visas being denied over the years.
df_denied = df[df['case_status'] == 'Denied']
df_denied
```

```python
df_denied.groupby('yr_of_estab').count().plot()
```

**The trend is that the visas being denied over the years have been increase**
