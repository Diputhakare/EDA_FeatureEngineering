*Z-Score*

```python
outlier=[]
def detect_outliers(data):
    threshold=3
    mean=np.mean(data["math score"])
    std=np.std(data["math score"])
    
    for i in data['math score']:
        z=(i-mean)/std
        if np.abs(z)> threshold:
            outlier.append(i)
    return outlier   
```

```python
outlier_pt=detect_outliers(data)
```

# Encoding

```python
data["lunch"].unique()
```

```python
#Ordinal Encoding:"Ordinal" means ordered
df=data.copy()
lunch_dict ={'standard':1,'free/reduced':0}
df['Ordinal_lunch']=df.lunch.map(lunch_dict)
df[['lunch','Ordinal_lunch']]
```

```python
#One-hot coding:Put the check mark '1' for the appropriate location.
df=data.copy()
df_type=pd.get_dummies(df[["race/ethnicity"]],prefix="R")

pd.concat([df[["race/ethnicity"]],df_type],axis=1).head()
```

```python
#binary Encoding
from category_encoders import BinaryEncoder
df=data.copy()

be=BinaryEncoder(cols=['race/ethnicity'])
newdata=be.fit_transform(df['race/ethnicity'])

EncounterStep= pd.DataFrame([1,2,3,1,1,3,2,3,4],columns=["EncounterStep"]) 
dfh=pd.concat([df[['race/ethnicity']],EncounterStep,newdata],axis=1)
dfh
```

```python
# Frequency Encoding:Give each category the probability (occurence/total event).
df=data.copy()

dfTemp=df.groupby("race/ethnicity").size()/len(df) #Group it by type, find the size of each type, and divide by total event
df['Type_freq']=df['race/ethnicity'].map(dfTemp) #dfTemp is a dataframe type

pd.concat([df[['race/ethnicity']],df['Type_freq']],axis=1)
```

```python
# Hashing Encoding
from sklearn.feature_extraction import FeatureHasher
df=data.copy()

fg = FeatureHasher(n_features=2, input_type='string')
hashed_features = fg.fit_transform(df['race/ethnicity'])
hashed_features = hashed_features.toarray()

df=pd.concat([df[['race/ethnicity']], pd.DataFrame(hashed_features)], axis=1)
df
```

```python
#Helmert (reverse) Encoding:Helmert coding compares each level of a categorical variable to the mean of the subsequent levels
from category_encoders import HelmertEncoder
df=data.sample(5)

he=HelmertEncoder(cols=['gender'])
newcolumn=he.fit_transform(df['gender'])

df=pd.concat([df[['gender']],newcolumn],axis=1)
df
```

```python
#Backward Difference Encoding
#In backward difference coding, the mean of the dependent variable for a level is compared with the mean of the dependent variable for the prior level
from category_encoders import BackwardDifferenceEncoder
df=data.sample(5)

bwde = BackwardDifferenceEncoder()
newcolumns=bwde.fit_transform(df['race/ethnicity'])

pd.concat([df[['race/ethnicity']],newcolumns],axis=1)
```

```python
 #Target Encoding
df=data.copy()

mean_encode=df.groupby("gender")['math score'].mean()
df['Type_race']=df['gender'].map(mean_encode)

df[['gender','math score','Type_race']]    
```

```python
# Leave One Out Encoding
#target encoding but excludes the current row’s target when calculating the mean target for a level to reduce the effect of outliers.
from category_encoders import LeaveOneOutEncoder
df=data.copy()

LOOE = LeaveOneOutEncoder(cols=['race/ethnicity'], sigma=0.2)
df['Type_legendary']=LOOE.fit_transform(df['race/ethnicity'], df['math score'])

df[['race/ethnicity','math score','Type_legendary']]
```

```python
#M-estimator Encoding
from category_encoders import MEstimateEncoder
df=data.copy()

MEE=MEstimateEncoder(m=2)
newcolumns = MEE.fit_transform(df['gender'], df['writing score'])

df['MEE_col']=newcolumns
df[['gender','writing score','MEE_col']]
```
